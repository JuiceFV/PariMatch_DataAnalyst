### Первый взгляд
Для начала я просто хочу взглянуть на корреляцию имеющихся данных

Я создал функцию, которая переводит categorical data в numerical data. Она приннимает в виде аргументов датасет и лист столбцов, которые надо дропнуть из результатирующего датасета. Функция возвращает полученный датасет и соответствия categorical и numerical data, например M - 1; F - 0.

```python
def categorical_to_numeric(dataset, drop=None):
    temp_df = dataset.copy()
    
    if drop:
        temp_df = temp_df.drop(drop, axis=1)
    
    for i in temp_df.select_dtypes(['object']).columns:
        temp_df[i] = temp_df[i].astype('category')
    
    cat_columns = temp_df.select_dtypes(['category']).columns

    
    source_relates = {}
    for i in cat_columns:
        source_relates.update({i: dict(enumerate(temp_df[i].cat.categories ))})
    
    temp_df[cat_columns] = temp_df[cat_columns].apply(lambda x: x.cat.codes)
    
    
    return temp_df, source_relates
```

```python
temp_df, rel_src = categorical_to_numeric(df, drop=['Id_client'])
```

Ну я бужу выбирать из тех способов нормализации, которые я знаю:
1. Gaussian nurmaliztion (сразу нет, т.к. данный способ применяется , если распределение фитч похоже на нормальное - в нашем случае не так)
2. Log Transform (так же не подходит, т.к. распределение фитч не подходит под [Power low](https://en.wikipedia.org/wiki/Power_law)
3. И пследнее - использование Квантилей. Ну этот метод используется в большинстве случаев и я им воспользуюсь. [Видео](https://www.youtube.com/watch?v=ecjN6Xpv6SE)

```python
def quantile_normalize(df):
    df_sorted = pd.DataFrame(np.sort(df.values,
                                     axis=0), 
                             index=df.index, 
                             columns=df.columns)
    df_mean = df_sorted.mean(axis=1)
    df_mean.index = np.arange(1, len(df_mean) + 1)
    df_qn =df.rank(method="min").stack().astype(int).map(df_mean).unstack()
    return(df_qn)
```

```python
normalized_temp_df = quantile_normalize(temp_df)
```

```python
fig, axes = plt.subplots(figsize=(16,9))
sns.boxplot(data=normalized_temp_df, ax=axes)
axes.xaxis.set_tick_params(rotation=90)
```

## Запускаем алгоритм

Алгоритм, который я выбрал - k-means, с его поиощью я выясню оптимальное кол-во класстеров. И для более флексабильных кластеров я использую Gaussian Mixture с полученным из k-means оптимальным кол-вом класстеров.

```python
normalized_temp_df
```

По elbow диаграме сложно что-то понять (но самый резкий переход/сгиб это в k=2), а следовательно воспользуемся Silhouette Score.

```python
distortions = []
K = range(1,50)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(normalized_temp_df)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Максимальная точка находится в k=2, что подтверждает предыдущее высказывание. А следовательно оптимальное кол-во класстеров - 2.

```python
sil = []
K = range(2,10)
for k in K:
    kmeans = KMeans(n_clusters = k).fit(normalized_temp_df)
    labels = kmeans.labels_
    sil.append(silhouette_score(normalized_temp_df, labels, metric = 'cosine'))
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('The Silhouette Score showing the optimal k')
plt.show()
```

```python
gmm = GaussianMixture(n_components=2, covariance_type='full', verbose=2).fit(normalized_temp_df)
```

```python
from sklearn import linear_model
import numpy as np
import scipy.stats as stat

class LogisticReg:
    def __init__(self,*args,**kwargs):#,**kwargs):
        self.model = linear_model.LogisticRegression(*args,**kwargs)#,**args)

    def fit(self,X,y):
        self.model.fit(X,y)
        # Get p-values for the fitted model
        denom = (2.0*(1.0+np.cosh(self.model.decision_function(X))))
        denom = np.tile(denom,(X.shape[1],1)).T
        F_ij = np.dot((X/denom).T,X) # Fisher Information Matrix
        Cramer_Rao = np.linalg.inv(F_ij) # Inverse Information Matrix
        sigma_estimates = np.sqrt(np.diagonal(Cramer_Rao))
        z_scores = self.model.coef_[0]/sigma_estimates # z-score for eaach model coefficient
        p_values = [stat.norm.sf(abs(x))*2 for x in z_scores] # two tailed test for p-values
        
        self.z_scores = z_scores
        self.p_values = p_values
        self.sigma_estimates = sigma_estimates
        self.F_ij = F_ij
```

```python
X_ct=check_df[['Count_transac']]
X_te=check_df[['Tenure']]
X_ta=check_df[['Total_amount']]
y=check_df.cluster
logit_ct = LogisticReg(C=1e5)
logit_ct.fit(X_ct,y)
logit_te = LogisticReg(C=1e5)
logit_te.fit(X_te,y)
logit_ta = LogisticReg(C=1e5)
logit_ta.fit(X_ta,y)
```

```python
from scipy.special import expit

fig, axes = plt.subplots(3, 1, figsize=(16,24))
X_test_ct = np.linspace(0, 1000, 10000)
X_test_te = np.linspace(0, 12, 10000)
X_test_ta = np.linspace(0, 22749, 10000)

axes[0].scatter(X_ct, y, color='black', zorder=20)
loss = expit(X_test_ct * logit_ct.model.coef_ + logit_ct.model.intercept_).ravel()
axes[0].plot(X_test_ct, loss, color='red', linewidth=3)

axes[1].scatter(X_te, y, color='black', zorder=20)
loss = expit(X_test_te * logit_te.model.coef_ + logit_te.model.intercept_).ravel()
axes[1].plot(X_test_te, loss, color='red', linewidth=3)

axes[2].scatter(X_ta, y, color='black', zorder=20)
loss = expit(X_test_ta * logit_ta.model.coef_ + logit_ta.model.intercept_).ravel()
axes[2].plot(X_test_ta, loss, color='red', linewidth=3)
```
